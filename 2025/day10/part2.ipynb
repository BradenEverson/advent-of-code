```python
import numpy as np
from scipy.optimize import milp, LinearConstraint
```

```python
with open("data/input", "r") as f:
    lines = f.readlines()
switches, targets = [], []
```

```python
for i in lines:
    j = i.strip().split(" ")[1:]
    targets.append([int(x) for x in j[-1][1:-1].split(",")])
    switches.append([[int(y) for y in x[1:-1].split(",")] for x in j[:-1]])

mats = []
for a,b in zip(switches, targets):
    xs = np.zeros((len(b), len(a)))
    for i in range(len(a)):
        xs[:, i][a[i]] = 1
    mats.append(xs)

total = 0
for A, b in zip(mats, targets):
    c = np.ones(A.shape[1])
    B = LinearConstraint(A, lb=b, ub=b)
    total += sum(milp(c, integrality=c, constraints=B).x)
print(total)
```
